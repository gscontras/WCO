```python
import pandas as pd
CSV_FILE_LOCATION = "/Users/katherine/Desktop/Maryland/WCO/WCO/experiments/1-rating/stimuli-gen/WCO stimuli break down.csv"
TXT_OR_JSON_FILE_SAVE_LOCATION = "/Users/katherine/Desktop/Maryland/WCO/WCO/experiments/1-rating/stimuli-gen/"
iDontKnowIfIAmSupposedToIncludeThisLineOfCode = False
```

```python
df = pd.read_csv(CSV_FILE_LOCATION)
```

```python
def program():    
    dataList = list()
    
    i = 0
    for l in df['Label']:
        dataList.append([l, "", "", ""])
        i += 1
    
    i = 0
    for cond in df['Condition']:
        dataList[i][1] = cond
        i += 1
        
    i = 0
    for s in df['Sentence example']:
        dataList[i][2] = s
        i += 1
        
    i = 0
    for cont in df['Context']:
        dataList[i][3] = cont
        i += 1
        
    for item in dataList:
        #item: [Label, Condition, Sentence, Context]
        label = item[0]
        condition = item[1]
        sentence = item[2]
        context = item[3]

    def functionOne(label: str, condition: str, sentence: str, context: str) -> str:
        returnStr = str()
        if iDontKnowIfIAmSupposedToIncludeThisLineOfCode:
            returnStr += "codeString"
            returnStr += condition
            returnStr += " = "

        returnStr += condition
        returnStr += ": {story: \""
        returnStr += context
        returnStr += "\", "

        returnStr += "sentence: \""
        returnStr += sentence
        returnStr += "\", "

        returnStr += "label: \""
        returnStr += label
        returnStr += "\"}"

        return returnStr

    def functionTwo(itemtype: str, label: str, dL: list) -> str:
        returnstr = label + "xconfig = \n"
        returnstr += "{itemtype: \""
        returnstr += itemtype
        returnstr += "\",\n\n"
        for item in dL:
            #item: [Label, Condition, Sentence, Context]
            if label == item[0]:
                returnstr += functionOne(item[0], item[1], item[2], item[3])
                returnstr += ",\n\n"
            
        #we don't want the extra comma/newlines at the end hence [:-3]
        returnstr = returnstr[:-3] + "\n}"
        
        fileLocation = TXT_OR_JSON_FILE_SAVE_LOCATION + label + "-expconfig.js"
        print(fileLocation)
        #print (fileLocation)
        #codeFile = open(fileLocation, 'w')
        #codeFile.write(returnstr)
        #codeFile.close()
        return "done! Yay!" + returnstr
    
    i = 0
    while i < len(dataList):
        #print(functionTwo(itemtype= 'test', label= dataList[i][0], dL= dataList))
        functionTwo(itemtype= 'test', label= dataList[i][0], dL= dataList)
        i += 8
        
    #print(functionTwo(itemtype= 'test', label= 'queen', dL= dataList))
    functionTwo(itemtype= 'test', label= 'queen', dL= dataList)

#if __name__ == '__main__':
program()
```
